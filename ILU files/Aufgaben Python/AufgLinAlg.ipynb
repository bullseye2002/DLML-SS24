## Vektor Multiplikation / Skalarprodukt / Inneres Produkt / Punktprodukt / Dot Product

* Skalarprodukt ergibt einen skalaren Wert (einzelne Zahl)

\begin{equation}x=\begin{bmatrix}3\\-1\\5\end{bmatrix}, \theta=\begin{bmatrix}2\\1.5\\-2\end{bmatrix}\end{equation}

\begin{equation}\theta \cdot x =\theta^T  x = 
\begin{bmatrix}\theta_0 &\theta_1&\cdots&\theta_{n-1}\end{bmatrix} 
\begin{bmatrix}x_0 \\ x_1\\\vdots\\x_{n-1}\end{bmatrix}=\sum_{i=0}^{n-1} \theta_i x_i
\end{equation}

\begin{equation} \begin{bmatrix}\color{blue}{2}&\color{green}{1.5}&\color{red}{-2}\end{bmatrix} \begin{bmatrix}\color{blue}{3}\\\color{green}{-1}\\\color{red}{5}\end{bmatrix}= \color{blue}{2\cdot 3}+\color{green}{1.5\cdot (-1)}+\color{red}{(-2)\cdot 5}=-5.5  \end{equation}
