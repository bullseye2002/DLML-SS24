$\color{blue}{\text{Aufgabe 3}}$. **Lineplot mit Konfidenzintervall**. Mit:

```python
x= np.arange(5)
y= np.matrix((x-2)**2).T; 
df = pd.DataFrame(np.tile(y,(1,3))+np.random.random_sample((len(x),3)))
df['xval'] = x
df
```

erstellen wir ein Pandas DataFrame, das in den Spalten 0, 1, 2 dreimal eine verrauschte Messung einer Parabel $(x-2)^2$ enthält. (Genausogut könnte in den Spalten auch drei Trainingskurven eines Deep-Learning-Modells stehen, die auch aufgrund der Zufallseffekte Schwankungen unterworfen sind). Stellen Sie einen zusammenfassenden Plot mit xval als x-Achse und den Spalten 0,1,2 als y-Daten in der Form "Kurve mit Konfidenzband" dar. 

Hilfestellung: Das gelingt in Seaborn besser, wenn wir die Daten zunächst in eine "zusammengeschmolzene" Form mit `pd.melt` bringen (siehe auch [https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf]):

```python
df2 = pd.melt(df,id_vars='xval')
df2
```

Hierbei wird die x-Variable `xval` mittels `ìd_vars='xval'` von der Verschmelzung in die neue Spalte `value` ausgenommen.
